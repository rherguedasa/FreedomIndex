```python
# hacemos una función para convertir los valores a string y además eliminar ciertos elementos ('$%')
def clean_and_convert_to_float(value):
    """ clean_and_convert_to_float convierte los valores a tipo float y quita str 
    :param value: valor a ser limpiado y convertido a float.
    :return: valor convertido a tipo float sin nigún simbolo
    """
    cleaned_value = ''.join(filter(str.isdigit, str(value))) # eliminamos todos los caracteres que no sean números o puntos decimales
    if not cleaned_value: # verificamos si la cadena resultante es vacía
        return None
    float_value = float(cleaned_value) # convertimos la cadena a un número float
    return float_value

# columnas a las que vamos a quitar los elementos que sobran
columns_to_clean = ['Population (Millions)', 'GDP (Billions, PPP)', 
                    'GDP per Capita (PPP)', 'Unemployment (%)', 'FDI Inflow (Millions)']

# aplicamos la función de limpieza y conversión a cada columna en la lista con un bucle for
for col in columns_to_clean:
    df[col] = df[col].apply(lambda x: clean_and_convert_to_float(x))
```

```python
# convertir estas columnas a float ['Population (Millions)', 'GDP (Billions, PPP)',	'GDP per Capita (PPP)',	'Unemployment (%)', 'FDI Inflow (Millions)']
df[['Population (Millions)', 'GDP (Billions, PPP)', 
    'GDP per Capita (PPP)', 
    'Unemployment (%)', 'FDI Inflow (Millions)']] = df[['Population (Millions)', 'GDP (Billions, PPP)', 
                                                        'GDP per Capita (PPP)', 'Unemployment (%)', 
                                                        'FDI Inflow (Millions)']].astype(float)
```

```python
# porcentaje de nulos 
percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing.sort_values(ascending = False)
```

```python
# sacamos un listado de las columnas con nulos
def columns_with_nulls(data):
    """columns_with_nulls recorre el df y hace un listado de las columnas con nulos
    :param data: dataframe
    :return: listado de columnas con nulos
    """
    columns_with_nulls = data.columns[data.isnull().any()].tolist()
    return columns_with_nulls
columns_with_nulls(df)
```

```python
# al tener pocos valores nulos, aplicamos una función que convierta los nulos en sus 5 KNN más cercanos
def fill_nulls_with_knn(df, columns_with_nulls):
    """fill_nulls_with_knn utiliza para reemplazar los valores nulos en un DataFrame con los 5 valores más cercanos según el algoritmo KNN.
    :param df: DataFrame que se desea modificar
    :columns_with_nulls: Lista de columnas en el DataFrame que contienen valores nulos.
    :return: nuevo DataFrame en el que se han reemplazado los valores nulos
    """
    df_filled = df.copy()
    for col in columns_with_nulls:
        imputer = KNNImputer(n_neighbors=5) 
        df_filled[col] = imputer.fit_transform(df_filled[[col]]) 
    return df_filled

# creamos la lista de columnas
columns_with_nulls = ['World Rank', 'Region Rank', '2019 Score', 'Property Rights', 
                    'Judical Effectiveness', 'Government Integrity', 'Tax Burden', "Gov't Spending",
                    'Fiscal Health', 'Business Freedom', 'Labor Freedom', 'Monetary Freedom',
                    'Trade Freedom', 'Investment Freedom ', 'Financial Freedom', 'Tariff Rate (%)',
                    'Income Tax Rate (%)', 'Corporate Tax Rate (%)', 'Tax Burden % of GDP',
                    "Gov't Expenditure % of GDP ", 'GDP (Billions, PPP)', 'GDP Growth Rate (%)',
                    '5 Year GDP Growth Rate (%)', 'GDP per Capita (PPP)', 'Unemployment (%)',
                    'Inflation (%)', 'FDI Inflow (Millions)','Public Debt (% of GDP)']

# aplicamos la función para nuestro df
df = fill_nulls_with_knn(df, columns_with_nulls)
```

```python
# eliminamos algunas columnas que no nos hacen falta y renombramos otras para su mejor lectura
df.drop(columns=['WEBNAME'], inplace=True)
df.drop(columns=['CountryID'], inplace=True)
df.drop(columns=['Country'], inplace=True)
df.rename(columns={'Country Name':'Country'}, inplace=True)
df.rename(columns={'GDP (Billions, PPP)':'GDP (Billions, PPP)$'}, inplace=True)
df.rename(columns={'GDP per Capita (PPP)':'GDP per Capita (PPP)$'}, inplace=True)
df.rename(columns={'Investment Freedom ':'Investment Freedom'}, inplace=True)
df.rename(columns={"Gov\'t Expenditure % of GDP ":"Gov\'t Expenditure % of GDP"}, inplace=True)
```

```python
# preprocesamos los detalles que nos encomtramos dentro del df coord 
def process_column(df, column_name):
    """process_column remplaza los signos guiones y los cambia por simbolo de restar
    :param df: simbolo de resta
    :column_name: nombre de la columna que se ve afectada
    :return: df modificado
    """
    df[column_name] = df[column_name].str.replace("–", "-")
    return df

# aplicamos la función
coord = process_column(coord, 'Latitude')
coord = process_column(coord, 'Longitude')
```

```python
# pasamos los valores a tipo float para poder luego graficar 
coord['Latitude'] = coord['Latitude'].astype(float)
coord['Longitude'] = coord['Longitude'].astype(float)
```
